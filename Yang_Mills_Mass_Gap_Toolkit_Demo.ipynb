This notebook demonstrates the synthetic spectral operator for the Yang–Mills mass gap problem.

The operator predicts the mass gap as a function of coupling strength g and generates black-hole-style entropy spectra.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the operator
tvals = np.load('yang_mills_operator_toolkit_strongest.npy')

def mass_gap_predictor(g):
    sorted_diag = np.sort(np.abs(tvals))
    gap = sorted_diag[1] if len(sorted_diag) > 1 else 0.0
    return gap / (1 + 0.8 * g**1.8) + 0.4

print('Operator loaded successfully!')
print('Example: mass gap at g=5.0 →', mass_gap_predictor(5.0))
```

```python
# Mass gap predictions
test_g = [1.2, 1.5, 2.0, 3.0, 5.0, 10.0]
for g in test_g:
    print(f'g = {g} → Predicted mass gap = {mass_gap_predictor(g):.3f}')
```

```python
# Black-hole entropy plot
beta = np.logspace(-2, 3, 300)
Z = np.sum(np.exp(-np.clip(beta[:, None] * tvals[:80000], -700, 700)), axis=1)
E_avg = np.sum(tvals[:80000] * np.exp(-np.clip(beta[:, None] * tvals[:80000], -700, 700)), axis=1) / (Z + 1e-12)
S = beta * E_avg + np.log(Z + 1e-12)

plt.figure(figsize=(9, 5))
plt.semilogx(beta, S, 'b-', label='Yang–Mills Spectral Entropy')
plt.axhline(0, color='r', linestyle='--', label='3rd law violation (like real BH)')
plt.xlabel('Inverse temperature β')
plt.ylabel('Entropy S(β)')
plt.title('Black-Hole Entropy from Yang–Mills Mass Gap Operator')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
